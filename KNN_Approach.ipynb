```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
numerical_features = df.select_dtypes(include=['float64', 'int64'])

imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(numerical_features)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Apply PCA
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)
```

```python
y = df['fund_sharpe_ratio_3years']

knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_pca, y)
```

```python
def recommend_funds_based_on_sharpe(target_sharpe_ratio, model, pca, scaler, df, top_n=5):
    # Predict Sharpe ratios for all funds
    predicted_sharpe_ratios = model.predict(X_pca)

    # Calculate distances from predicted Sharpe ratios to the target Sharpe ratio
    distances = np.abs(predicted_sharpe_ratios - target_sharpe_ratio)


    top_indices = np.argsort(distances)[:top_n]


    return df.iloc[top_indices]


target_sharpe_ratio = 1.0

# Get recommendations
recommended_funds = recommend_funds_based_on_sharpe(target_sharpe_ratio, knn, pca, scaler, df)
print(recommended_funds)
```

```python
X_train, X_test = X_pca, X_pca
y_train, y_test = y, y

knn.fit(X_train, y_train)

predicted_sharpe_ratios = knn.predict(X_test)

mae = mean_absolute_error(y_test, predicted_sharpe_ratios)
rmse = np.sqrt(mean_squared_error(y_test, predicted_sharpe_ratios))
accuracy_score = knn.score(X_test, y_test)
mse=mean_squared_error(y_test, predicted_sharpe_ratios)
print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")
print(f"Mean Squared Error: {mse}")
print(f"Accuracy Score: {accuracy_score}")
```

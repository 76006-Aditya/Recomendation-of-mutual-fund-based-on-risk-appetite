```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
Mutual_Funds['size_type_encoded'] = label_encoder.fit_transform(Mutual_Funds['size_type'])
Mutual_Funds['investment_type_encoded'] = label_encoder.fit_transform(Mutual_Funds['investment_type'])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
def determine_risk_level(row):
    # now we will define the risk levels based on shapiro and aplha
    if row['fund_sharpe_ratio_3years'] > 1.5 and row['fund_alpha_3years'] > 0.03:
        return 'High'
    elif row['fund_sharpe_ratio_3years'] > 1.0 and row['fund_alpha_3years'] > 0.01:
        return 'Medium'
    else:
        return 'Low'

Mutual_Funds['risk_level'] = Mutual_Funds.apply(determine_risk_level, axis=1)
```

```python
train_df, test_df = train_test_split(Mutual_Funds, test_size=0.4, random_state=42)


scaler = StandardScaler()
features = ['fund_sharpe_ratio_3years', 'fund_alpha_3years', 'fund_return_3years',
             'fund_sharpe_ratio_5years', 'fund_alpha_5years', 'fund_return_5years']

train_df[features] = scaler.fit_transform(train_df[features])
test_df[features] = scaler.transform(test_df[features])


risk_metrics_train = train_df[features]
similarity_matrix = cosine_similarity(risk_metrics_train)
similarity_df = pd.DataFrame(similarity_matrix, index=train_df['fund_symbol'], columns=train_df['fund_symbol'])


def recommend_funds(input_risk_level, top_n=3):

    filtered_df = Mutual_Funds[Mutual_Funds['risk_level'] == input_risk_level]

    if filtered_df.empty:
        raise ValueError("No funds available for the given risk level.")


    risk_metrics_filtered = filtered_df[features]
    similarity_matrix_filtered = cosine_similarity(risk_metrics_filtered)
    similarity_df_filtered = pd.DataFrame(similarity_matrix_filtered, index=filtered_df['fund_symbol'], columns=filtered_df['fund_symbol'])


    recommendations = []
    for fund in filtered_df['fund_symbol']:
        sim_scores = similarity_df_filtered[fund].drop(fund)  # Exclude self
        top_similar_funds = sim_scores.nlargest(top_n).index
        recommendations.extend(top_similar_funds)

    # Remove duplicates and select top N
    recommendations = list(set(recommendations))
    return recommendations[:top_n]
```

```python
def evaluate_recommendations(input_risk_level, top_n=3):
    recommendations = recommend_funds(input_risk_level, top_n)


    relevant_funds = test_df[test_df['risk_level'] == input_risk_level]['fund_symbol'].tolist()


    y_true = [1 if fund in relevant_funds else 0 for fund in recommendations]
    y_pred = [1] * len(recommendations)  # As recommendations are considered positive


    precision = precision_score([1] * len(y_true), y_true, zero_division=0)
    recall = recall_score([1] * len(y_true), y_true, zero_division=0)
    f1 = f1_score([1] * len(y_true), y_true, zero_division=0)

    return precision, recall, f1

# Take user input for risk level
input_risk_level = input("Enter the risk level (High, Medium, Low) to get recommendations: ")
top_n = int(input("Enter the number of top similar funds to recommend: "))

try:

    recommended_funds = recommend_funds(input_risk_level, top_n)
    print(f"Recommended funds for risk level {input_risk_level}: {recommended_funds}")


    precision, recall, f1 = evaluate_recommendations(input_risk_level, top_n)
    print(f"Evaluation Metrics for risk level {input_risk_level}:")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")

except ValueError as e:
    print(e)
```
